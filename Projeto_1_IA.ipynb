```python
colunas_analise = ['len_AStar', 'dist_AStar', 'time_AStar',
                   'len_dfs', 'dist_dfs', 'time_dfs',
                   'delta_time', 'delta_len', 'delta_dist']

# Proporções A* e Profundidade
df_estat = df_obs[colunas_analise].copy()
df_estat['len_rate'] = df_obs["len_AStar"]/df_obs["len_dfs"]
df_estat['dist_rate'] = df_obs["dist_AStar"]/df_obs["dist_dfs"]
df_estat['time_rate'] = df_obs["time_AStar"]/df_obs["time_dfs"]

# Comparação A* e Profundidade
df_comp = pd.DataFrame({'caminhos_iguais' : [abs(len(df_estat.loc[df_estat['delta_dist']==0])/len(df_estat))],
                        'delta_tamanhos_medio' : [abs(df_estat['delta_len'].mean())],
                        'delta_tamanhos_max' : [abs(df_estat['delta_len']).max()],
                        'delta_distancia_medio' : [abs(df_estat['delta_dist'].mean())],
                        'delta_distancia_max' : [abs(df_estat['delta_dist']).max()],
                        'delta_tempo_medio' : [abs(df_estat['delta_time'].mean())],
                        'delta_tempo_max' : [abs(df_estat['delta_time']).max()],
                        })

fig, axs = plt.subplots(1,2, figsize=(12,6))

list_var = [[['len_', 'dist_'], ['Tamanho', 'Distância'], ['red', 'blue']],
            [['time_'], ['Tempo'], ['green']]]
list_title = ['Tamanho e Distância', 'Tempo']

for i, plot in enumerate(list_var):
    ax = axs[i]
    for var in range(len(plot[i])):
        sns.histplot(df_estat[plot[0][var] + 'rate'].loc[df_estat[plot[0][var] + 'rate'] != 1],
                     bins=250, kde=True, alpha = 0.5, label = plot[1][var], color = plot[2][var], ax=ax)
        
    ax.set_xlabel(r'Proporção $(\frac{A*}{Depth})$')
    ax.set_ylabel('Número de caminhos')
    ax.set_title('Proporção entre resultados - ' + list_title[i])
    ax.legend()

plt.tight_layout()
plt.savefig("Proporcoes_buscas.png")
plt.show()
```
